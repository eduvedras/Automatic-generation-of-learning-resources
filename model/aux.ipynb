```python
import pandas as pd
from tqdm import tqdm
filetag = 'gpt3.5-finetuned'
data = pd.read_csv(f'./results-final/results/{filetag}.csv')

question_dataset = pd.read_csv('../template_generation/QuestionsDataset.csv',sep=';')

references = []
predictions = []
for index,row in tqdm(data.iterrows()):
    #questions = row['References']
    predictions.append(row['Prediction'])
    references_aux = []
    for indexq, rowq in question_dataset.iterrows():
        if rowq['Chart'] == row['Image']:
            references_aux.append(rowq['Question'])
    references.append(references_aux)
    #references.append(questions[2:-2].split('\', \''))

references
```

```python
import evaluate

file = open(f"./results-final/not_used/{filetag}.txt", "a")

bleu = evaluate.load("bleu")
results = bleu.compute(predictions=predictions, references=references)
print(results)
file.write(f"BLEU: {results}\n")

meteor = evaluate.load("meteor")
results = meteor.compute(predictions=predictions, references=references)
print(results)
file.write(f"METEOR: {results}\n")

rouge = evaluate.load("rouge")
results = rouge.compute(predictions=predictions, references=references)
print(results)
file.write(f"ROUGE: {results}\n")
file.close()
```
